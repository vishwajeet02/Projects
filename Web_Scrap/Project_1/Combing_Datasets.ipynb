```python
df = pd.DataFrame()

df['title']         = df_1['title']
df['upc']           = df_2['upc']
df['category']      = df_2['typ']
df['price_e_tax']   = df_2['price_e']
df['price_i_tax']   = df_2['price_i']
df['tax']           = df_2['tax']
df['rating']        = df_1['rating']
df['reviews']       = df_2['no_of_reviews']
df['quantity']      = df_2['Quantity']
df['stock']         = df_1['stock']

df['book_link']     = df_1['book_link']
df['img_link']      = df_1['img_link']
```
