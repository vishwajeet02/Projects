```python
import requests
import pandas as pd
from tqdm import tqdm
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv('Book_Scraper_Final.csv')
df.head()
```

```python
data =[]
for link in tqdm(df['book_link']):
    res = requests.get(link)
    soup = BeautifulSoup(res.text,'html.parser')
    typ = soup.find('ul',class_='breadcrumb').find_all('a')[2].text
    temp = soup.find('table',class_ = 'table table-striped').find_all('td')

    upc = temp[0].text
    price_e = temp[2].text[2:]
    price_i = temp[3].text[2:]
    tax = temp[4].text[2:]
    Availability = temp[5].text
    no_of_reviews = temp[6].text

    
    data.append([typ,price_e,price_i,tax,Availability,no_of_reviews,upc])    
```

```python
df = pd.DataFrame(data,columns = ['typ','price_e','price_i','tax','Quantity','no_of_reviews','upc'])
df.head()
```

```python
df.to_csv('Individual_Scrap.csv',index= False)
```
